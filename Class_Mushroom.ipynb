```
train = pd.read_csv('train_mush.csv')
test = pd.read_csv('test_mush.csv')
```

```
train.info() # 23 Attributes ,? = missing value, objectหมดเลย
```

```
train.head() 
```

```
train.drop('stalk-root',axis=1,inplace = True) #lot of missing value --> dropout
test.drop('stalk-root',axis=1,inplace = True)
```

```
from sklearn.preprocessing import LabelEncoder #convert to numeric
X_train = train.copy()
X_test = test.copy()
for col in X_train.columns.values:
    train[col] = LabelEncoder().fit_transform(train[col])

for col2 in X_test.columns.values:
    test[col2] = LabelEncoder().fit_transform(test[col2])
Y_train = X_train['class'] 
Y_test = X_test['class']
```

```
from sklearn.model_selection import StratifiedKFold, cross_val_score 
from sklearn.metrics import roc_auc_score, roc_curve, auc, confusion_matrix
```

```
from sklearn.ensemble import RandomForestClassifier
RDF = RandomForestClassifier()
RDF.fit(X_train, Y_train)
```

```
Y_pred = RDF.predict(X_test) 
```

```
CM = confusion_matrix(Y_test,Y_pred)
CM
```

```
RDF.score(X_test,Y_test)
```
